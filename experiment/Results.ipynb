```python
import pandas as pd
import plotly.express as px
```

```python
results_wide = pd.read_csv("/content/drive/MyDrive/Tue/Seminars/Results wide - Sheet1.csv")
results_long = pd.read_csv("/content/drive/MyDrive/Tue/Seminars/Results long - Sheet1.csv")
```

```python
results_wide.head(100)
```

```python
results_long.head()
```

```python
accuracy_df = results_long.groupby(['study','log','algorithm'])['accuracy'].agg(value='mean').reset_index()
f1_df = results_long.groupby(['study','log','algorithm'])['f1'].agg(value='mean').reset_index()
auc_df = results_long.groupby(['study','log','algorithm'])['auc'].agg(value='mean').reset_index()

accuracy_df.head()
```

```python


fig = px.bar(accuracy_df, x="algorithm", y="value",
             color='study', barmode='group', facet_col="log",
             height=400, title="Accuracy")
fig.update_yaxes(range=[0, 1])
fig.show()
```

```python
fig = px.bar(f1_df, x="algorithm", y="value",
             color='study', barmode='group', facet_col="log",
             height=400, title="F1-score")
fig.update_yaxes(range=[0, 1])
fig.show()
```

```python
fig = px.bar(auc_df, x="algorithm", y="value",
             color='study', barmode='group', facet_col="log",
             height=400, title="AUC (ROC)")
fig.update_yaxes(range=[0, 1])
fig.show()
```
